```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
```

```python
train_df = pd.read_csv('datasets/clean/processedtrain.csv')
test_df = pd.read_csv('datasets/clean/processedtest.csv')
```

```python
#label encoding 
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

train_df['NObeyesdad']=encoder.fit_transform(train_df['NObeyesdad'])
train_df['NObeyesdad'].value_counts()

test_df['NObeyesdad']=encoder.fit_transform(test_df['NObeyesdad'])
test_df.head(11)
```

```python
dataset=['Weight','Height','Age','FCVC','FAVC','TUE']
X_train = train_df[dataset]
y_train = train_df['NObeyesdad']

X_test = test_df[dataset]
y_test = test_df['NObeyesdad']
```

```python
model = XGBClassifier(max_depth=3,               
    min_child_weight=5,                        
    subsample=0.8,            
    colsample_bytree=0.8,      
    learning_rate=0.05,       
    n_estimators=85,          
    random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
train_acc= accuracy_score(y_train,y_train_pred)
print(f"train Accuracy: {train_acc:.2f}")
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
