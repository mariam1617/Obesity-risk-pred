```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
train_df = pd.read_csv(r'C:\Users\Dell\Downloads\Ai_Project\Obesity-risk-pred\datasets\clean\processedtrain.csv')
test_df = pd.read_csv(r'C:\Users\Dell\Downloads\Ai_Project\Obesity-risk-pred\datasets\clean\processedtrain.csv')
```

```python
#label encoding 
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

train_df['NObeyesdad']=encoder.fit_transform(train_df['NObeyesdad'])
train_df['NObeyesdad'].value_counts()

test_df['NObeyesdad']=encoder.fit_transform(test_df['NObeyesdad'])
test_df.head(11)
```

```python
dataset=['Weight','Height','Age','FAVC','FAF','TUE']
X_train = train_df[dataset]
y_train = train_df['NObeyesdad']

X_test = test_df[dataset]
y_test = test_df['NObeyesdad']
```

```python
model = XGBClassifier(n_estimators=20, learning_rate=0.1, max_depth=5, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
