## Chi-Square Test

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import LabelEncoder

cat_cols = []
num_cols = []

for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        num_cols.append(column)
    elif pd.api.types.is_object_dtype(df[column]):
        cat_cols.append(column)
        

# Encode categorical variables (chi2 requires numeric inputs)
le = LabelEncoder()
X_encoded = df.apply(le.fit_transform)
y_encoded = df.apply(le.fit_transform)

X_cat =X_encoded[cat_cols].iloc[:, :-1]  # All categorical features except last column
y_cat = y_encoded[cat_cols].iloc[:, -1]    # Last categorical column as target
# Feature selection
selector = SelectKBest(chi2, k=5)  # Ensure k <= number of features
X_new = selector.fit_transform(X_cat, y_cat)

# Get selected feature names
selected_features = X_cat.columns[selector.get_support()]
print("Selected features:", selected_features)
```

```python
# Get chi-square scores and p-values
chi_scores, p_values = chi2(X_cat, y_cat)

chi_df = pd.DataFrame({
    'Feature': X_cat.columns,
    'Chi-Square': chi_scores,
    'p-value': p_values
}).sort_values('Chi-Square', ascending=False)


plt.figure(figsize=(12, 6))
bars = plt.barh(chi_df['Feature'], chi_df['Chi-Square'])
plt.xlabel('Chi-Square Score')
plt.title('Feature Importance ')
```

## Mutual Information

```python

from sklearn.feature_selection import mutual_info_classif

mut_cat=df.drop('NObeyesdad', axis=1)
mut_target=df.NObeyesdad
X_encoded = mut_cat.apply(lambda col: le.fit_transform(col))

# Encode target
y_encoded = le.fit_transform(mut_target)


# For categorical features
mi_scores = mutual_info_classif(X_encoded, y_encoded, discrete_features=True)
```

```python
mi_df = pd.DataFrame({
    'Feature': mut_cat.columns,
    'MI_Score': mi_scores
}).sort_values('MI_Score', ascending=False)


plt.figure(figsize=(12, 6))
sns.barplot(x='MI_Score', y='Feature', data=mi_df)

plt.title('Mutual Information Scores for Categorical Features')
plt.xlabel('Mutual Information Score')
plt.ylabel('Features')
plt.tight_layout()
plt.show()
```
